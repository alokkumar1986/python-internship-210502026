```python
s = {1, 2, 3, 4, 5 }
s1 = {4, 5, 6, 7, 8}
s2 = s.union(s1)
print(s)
print(s1)
print(s2)

s2 = s | s1
print(s2)
```

```python
s = {1, 2, 3, 4, 5 }
s1 = {4, 5, 6, 7, 8}
s2 = s.intersection(s1)
print(s)
print(s1)
print(s2)

s2 = s & s1
print(s2)
```

```python
s = {1, 2, 3, 4, 5 }
s1 = {4, 5, 6, 7, 8}
s2 = s.difference(s1)
print(s)
print(s1)
print(s2)

s2 = s - s1
print(s2)
```

```python
s = {1, 2, 3, 4, 5 }
s1 = {4, 5, 6, 7, 8}
s.difference_update(s1)

#s = s - s1

print(s)
```

```python
s = {1, 2, 3, 4, 5 }
s1 = {4, 5, 6, 7, 8}
s2 = s.symmetric_difference(s1)
print(s)
print(s1)
print(s2)

s2 = s ^ s1
print(s2)
```

```python
s = {1, 2, 3, 4, 5 }
s1 = {4, 5, 6, 7, 8}
s.symmetric_difference_update(s1)

#s ^ s = {}
#s ^ s1 = {1, 2, 3, 6, 7, 8}
print(s)
```

```python
s = {1, 2, 3, 4, 5, 6, 7}
s1 = {3, 4, 5}
x = {'A', "B", "C"}

print(s1.issubset(s))
print(s.issuperset(s1))
print(s1.isdisjoint(x))
```
