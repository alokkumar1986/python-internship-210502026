```python
#create rectangle class with three perporties width, height and findArea().
#Find the area of the rectange using this class.

class Rectangle:
  def __init__(self, width, height):
    self.width = width
    self.height = height
  def findArea(self):
    return self.width*self.height

r = Rectangle(15, 20)
print("Area = ", r.findArea())
```

```python
#Create Countdown class to print timer countdown
import time

class Countdown:
   def __init__(self, num):
     self.num = num
   def showTimer(self):
     while(self.num>=0):
      time.sleep(1)
      print('\r', self.num, end="")
      self.num = self.num-1
c = Countdown(10)
print("Timer : ", end="")
print("")
c.showTimer()
```
