```python
class Person:
  name = 'Aptech'
  age = 10
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def showName(self):
    print(self.name)

  @classmethod
  def showName1(cls):
    print(cls.name)

  @staticmethod
  def showAge():
    print(Person.age)


p = Person("Ardhendu", 31)
p.showName()
Person.showName1()
p.showAge()
```

```python
class Person:

  def __init__(self, name, age):
    self.name = name
    self.__age = age #private

  def getAge(self):
     return self.__age

  def setAge(self, age):
    self.__age = age


p = Person("Ardhendu", 31)
print(p.name)
p.setAge(32)
print(p.getAge())
```

```python
class Person:

  def __init__(self, name, age):
    self.name = name
    self._age = age #protected

p = Person("Ardhendu", 31)
print(p.name)
print(p._age)
```

```python
#write a program to find square of a number using private specifier of a class.

class Numb:
  def __init__(self, num):
    self.__num =  num
  def findSquare(self):
    return self.__num**2

n = Numb(10)
print(n.findSquare())
```

```python
class Numb:
  def __init__(self, num):
    self.__num =  num
  def __findSquare(self):
    return self.__num**2
  def getSquare(self):
    return self.__findSquare()

n = Numb(10)
print(n.getSquare())
```

```python
class Numb:
  def __init__(self, num):
    self.__num =  num
  def __findSquare(self):
    return self.__num**2


n = Numb(10)
print(n._Numb__num)
print(n._Numb__findSquare())
```

```python
class Vehicle:

  def __init__(self, name, color, price):
    self.name = name
    self.color = color
    self.price = price
  def show(self):
    print("Name:", self.name, "Color:", self.color, "Price:", self.price)


class Car(Vehicle):
    pass


c = Car("Creta", "Black", 2000000)
v = Vehicle("Creta", "Black", 2000000)
c.show()
v.show()
```

```python
class Vehicle:
  def __init__(self, name):
     self.name= name

  def show(self):
    print(self.name, "Drive")

class Car(Vehicle):
    def __init__(self, name, color):
      super().__init__(name)
      self.color= color

    def show(self):
      print(self.name)
      print(self.color, "Drive Car")

class Boat(Car):
     def show(self):   #method Overiding (Polymorphism)
       print(self.name, "Sails")

b = Car("Creta", "black")
b.show()
```

```python
#write a program to find area of three class triangle, square, rectangle
#with method overriding and inheritance.

class Shape:
  def __init__(self, l, h):
    self.l = l
    self.h = h
  def area(self):
    print("Area of shape")
class Trianlge(Shape):
  def area(self):
    return self.l*self.h/2
class Square(Shape):
  def area(self):
    return self.l*self.h
class Rectangle(Square):
    pass

r = Rectangle(10, 20)
print(r.area())
t = Trianlge(10, 20)
print(t.area())
s = Square(10, 10)
print(s.area())
```
