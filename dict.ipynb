```python
#write a program to constrct a dict with keys as numbers from 2 to 10
#and values as square of the key

d = {x : x*x for x in range(2, 11)}
print(d)

#===============================================

d = {}
for x in range(2, 11):
  d.update({ x : x*x})

print(d)
```

```python
#write a program to construct a dict with keys 2 to 11 (which are even)
#and values as cube of key

d = {x : x*x*x for x in range(2, 11) if x % 2 == 0}
print(d)


d = {}
for i in range(2, 11):
  if i % 2 == 0:
    d.update({i : i*i*i})

print(d)
```

```python
#write a program to swap keys and values of the dict
#d = {2: 8, 4: 64, 6: 216, 8: 512, 10: 1000}
d = {2: 8, 4: 64, 6: 216, 8: 512, 10: 1000}
d1 = {value : key for key, value in d.items()}
print(d1)
#=============================================
d1 = {}
for key, value in d.items():
  d1.update({value : key})
print(d1)
```
