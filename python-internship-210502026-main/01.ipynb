## Prime Numbers between 1 and 20

```python
prime_numbers = []
for num in range(2, 21):  # Start from 2 as 1 is not prime
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if (num % i) == 0:
            is_prime = False
            break
    if is_prime:
        prime_numbers.append(num)

print(f"Prime numbers between 1 and 20 are: {prime_numbers}")
```
